#### Changing gender to 1/0 
#### 0: Female
#### 1: Male

```python
Gender = pd.get_dummies(df['Gender'], drop_first=True).astype(int)
```

```python
Gender
```

```python
df = pd.concat([df, Gender], axis = 1)
```

```python
df.head()
```

```python
df.drop('Gender', axis=1, inplace=True)
```

```python
df.head(4)
```

## Normalizing time

```python
# Convert time columns to datetime format
df['Bedtime'] = pd.to_datetime(df['Bedtime'])  # Adjust format as needed
df['Wakeup time'] = pd.to_datetime(df['Wakeup time']) 
```

```python
# Extract hours and minutes from the time columns
df['Bedtime_hour'] = df['Bedtime'].dt.hour + df['Bedtime'].dt.minute / 60
df['Wakeup_hour'] = df['Wakeup time'].dt.hour + df['Wakeup time'].dt.minute / 60

# Normalize times to 0-1 scale (e.g., divide by 24 hours)
df['Normalized_Bedtime'] = df['Bedtime_hour'] / 24
df['Normalized_Wakeup'] = df['Wakeup_hour'] / 24
```

```python
df.head()
```

```python
df.drop(['Bedtime_hour', 'Wakeup_hour', 'Bedtime', 'Wakeup time'], axis=1, inplace=True)
```

## Data splitting and scaling

```python
df.drop('ID', axis=1, inplace=True)
```

```python
x=df.drop('Sleep efficiency', axis=1)
y=df['Sleep efficiency']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.4)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaled_x_train = scaler.fit_transform(x_train, y_train)
scaled_x_test = scaler.transform(x_test)
```

## Model Training - Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dec_tree = DecisionTreeRegressor()
```

```python
dec_tree.fit(scaled_x_train, y_train)
```

```python
dec_tree_pred = dec_tree.predict(scaled_x_test)
```

```python
dec_tree_pred
```

```python
np.array(y_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test, dec_tree_pred)
print(f'Mean Squared Error: {mse}')
```

```python
accuracy = dec_tree.score(scaled_x_test, y_test)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
y_pred_labels = (dec_tree_pred >= 0.5).astype(int)
y_test_labels = (y_test >= 0.5).astype(int)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test_labels, y_pred_labels)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

## Model Training - Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
random_forest = RandomForestRegressor(1000)
random_forest.fit(scaled_x_train, y_train)
```

```python
random_forest_pred = random_forest.predict(scaled_x_test)
```

```python
random_forest_pred
```

```python
y_test
```

```python
accuracy = random_forest.score(scaled_x_test, y_test)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

## Model Training-Support vector machine

```python
from sklearn.svm import SVR
supp = SVR()
supp.fit(scaled_x_train, y_train)
```

```python
supp_pred = supp.predict(scaled_x_test)
```

```python
accuracy = supp.score(scaled_x_test,y_test)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
lin = linear.score(scaled_x_test,y_test)
print(f"Multiple Linear Regression: {lin * 100:.2f}%")
dec = dec_tree.score(scaled_x_test,y_test)
print(f"Decision Tree: {dec * 100:.2f}%")
ran = random_forest.score(scaled_x_test,y_test)
print(f"Random Forest: {ran * 100:.2f}%")
svc = supp.score(scaled_x_test,y_test)
print(f"Support Vector Machine: {svc * 100:.2f}%")
```
